My first equation is $y^3 + 3x^2y-6x^2-6y^2+2$. 

We can find the local min/max by setting the first partial derivative equal to zero to find the critical points and then using the second partial derivative test to determine if the critical points are local min/max. 

For this equation, the critical point that obtains a local max, $p_1$, is $(0,0)$ 


```
from mpl_toolkits import mplot3d

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt



def f(x, y):
    return (y ** 3) + ((3 * x ** 2) * y) - (6 * x ** 2) - (6 * y ** 2) + 2

x = np.linspace(-1, 1, 30)
y = np.linspace(-1, 1, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(30, 35)
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.text2D(-0.035, 0.055, "This is a local maximum", color="red", size=15)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

```

My second equation is $\sqrt{16 - x^2 - y^2}$, which is the upper half of a spherical surface. 

We can find the local min/max by setting the first partial derivative equal to zero to find the critical points and then using the second partial derivative test to determine if the critical points are local min/max. 

For this equation, the critical point that obtains a local min, $p_2$, is $(0,0)$ 

```
from mpl_toolkits import mplot3d

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt



def f(x, y):
    return -np.sqrt(16 - x ** 2 - y ** 2)

x = np.linspace(-1, 1, 30)
y = np.linspace(-1, 1, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(60, 30)
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.text2D(-0.0375, -0.055, "This is a local minimum", color="red", size=13)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

My second equation is $y^2 - x^2$, which is a hyperbolic paraboloid. 

We can find the local min/max by setting the first partial derivative equal to zero to find the critical points and then using the second partial derivative test to determine if the critical points are local min/max. 

For this equation, the critical point that obtains a saddle point, which is neither a min nor a max, $p_3$, is $(0,0)$ 

```
from mpl_toolkits import mplot3d

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt



def f(x, y):
    return y ** 2 - x ** 2

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(30, 30)
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.text2D(-0.025, 0.035, "This is a saddle point", color="red", size=15)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

```

```
import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

# generate m random pairs (x, y)

m = 150
X = 2 * np.random.rand(m, 1)
X2 = 2 * np.random.rand(m, 1)
y = 4 + 3 * X + 2 * X2 + np.random.rand(m, 1)

# Show our plot
plt.plot(X + X2, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 3, 0, 15])
plt.show()
```

## Combine bias $b=w_0$ and weights $w_1$ and $w_2$

```
# add x0 = 1 
X_b = np.column_stack([np.ones((m, 1)), X, X2])
# create array with x1=0 and x1=2
X_new = np.array([[0], [2]])

# create array with x2 = 3 and x2 = 3
X_new2 = np.array([[0], [3]])

# add x0 = 1 to each instance and combine the two new arrays
X_new_b = np.column_stack([np.ones((2, 1)), X_new, X_new2])

# number of epochs
epochs = 20

# learning rate
lr = 0.01

# fix initial random weight for gradient descent (batch gradient descent)
np.random.seed(42)
initial_weight1 = np.random.randn(3, 1)
initial_weight2 = np.random.randn(3, 1) 

# Set weight as our two inital weights multiplied together
weight = initial_weight1.dot(initial_weight2.T)
weight_path_mgd = []
batch_size = 4

# plot initial prediction
y_predict = X_new_b.dot(weight)
plt.plot(X_new, y_predict, "r--")

weight_path_mgd.append(weight)
for epoch in range(epochs):
    shuffled_indices = np.random.permutation(m)
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    for i in range(0, m, batch_size):
        xi = X_b_shuffled[i:i+batch_size]
        yi = y_shuffled[i:i+batch_size]
        gradient = 1 / batch_size * xi.T.dot(xi.dot(weight) - yi)
        weight = weight - lr * gradient
        weight_path_mgd.append(weight)
        y_predict = X_new_b.dot(weight) 
        plt.plot(X_new, y_predict, "b-")

```
